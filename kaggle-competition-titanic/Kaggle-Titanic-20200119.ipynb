```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train['Survived'] = train['Survived'].astype(int)
train.info()
train.head(5)
```

```python
test.head(5)
```

```python
titanic_drop = pd.concat([train,test],sort=False)
```

```python
#drop_features = ['PassengerId','Survived']
#titanic_drop = titanic.drop(drop_features,axis=1)
titanic_drop.head()
```

```python
titanic_drop.info()
```

```python
titanic_drop['Title'] = titanic_drop.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())
```

```python
titanic_drop.head()
```

```python
titanic_drop.Title.value_counts()
```

```python
list1 = titanic_drop.Title.value_counts().index[4:].tolist()
mapping = {}

for i in list1:
    mapping[i] = 'Rare'
titanic_drop['Title'] = titanic_drop['Title'].map(lambda x: mapping[x] if x in mapping else x)
titanic_drop.Title.value_counts()
titanic_drop.to_csv('titanic_drop.csv',index=False)
```

```python
# 整合sibsp和parch
titanic_drop['FamilySize'] = titanic_drop.Parch + titanic_drop.SibSp + 1
titanic_drop=titanic_drop.drop(['SibSp','Parch'],axis=1)

titanic_drop['concat'] = titanic_drop['FamilySize'].map(str)+titanic_drop['Title']
```

```python
titanic_drop.concat.value_counts()
```

```python
grouped = titanic_drop.groupby(['concat'])
median = grouped.Age.median()
median
```

```python
#补全Age

def newage(cols):
    age = cols[0]
    concat = cols[1]
    if pd.isnull(age):
        return median[concat]
    return age

titanic_drop.Age = titanic_drop[['Age','concat']].apply(newage,axis=1)
```

```python
titanic_drop.info()
```

```python
grouped2 = titanic_drop.groupby(['Title'])
median2 = grouped2.Age.median()
#补全Age2

def newage2(cols):
    age = cols[0]
    title = cols[1]
    if pd.isnull(age):
        return median2[title]
    return age

titanic_drop.Age = titanic_drop[['Age','Title']].apply(newage2,axis=1)
titanic_drop.info()
```

```python
titanic_drop.Cabin = titanic_drop.Cabin.fillna('U')
titanic_drop[:5]
```

```python
# 补全Embarked
most_embarked = titanic_drop.Embarked.value_counts().index[0]
titanic_drop.Embarked=titanic_drop.Embarked.fillna('C')
titanic_drop.info()
```

```python
#补全Fare
titanic_drop.Fare = titanic_drop.Fare.fillna(titanic_drop.Fare.median())
titanic_drop.info()
```

```python
titanic_drop['Cabin'] = titanic_drop.Cabin.apply(lambda cabin:cabin[0])
titanic_drop.Cabin.value_counts()
```

```python
titanic_drop['Cabin'].loc[titanic_drop.Cabin=='T']='G'
titanic_drop.Cabin.value_counts()
```

```python
#按照年龄段将Age转化为类别特征
titanic_drop.Age=pd.cut(titanic_drop.Age,8,labels=False)
titanic_drop.head()
```

```python
titanic_drop = titanic_drop.drop('Name',axis=1)
titanic_drop = titanic_drop.drop('Ticket',axis=1)
titanic_drop = titanic_drop.drop('concat',axis=1)
```

```python
#按照Fare段将Fare转化为类别特征
titanic_drop.Fare=pd.cut(titanic_drop.Fare,4,labels=False)
titanic_drop.head()
```

```python
titanic_drop.Sex=titanic_drop.Sex.map({'male':1,'female':0})
titanic_drop.Cabin=titanic_drop.Cabin.map({'A':0,'B':1,'C':2,'D':3,'E':4,'F':5,'G':6,'U':7})
titanic_drop.Embarked=titanic_drop.Embarked.map({'C':0,'Q':1,'S':2})
titanic_drop.Title=titanic_drop.Title.map({'Mr':0,'Miss':1,'Mrs':2,'Master':3,'Rare':4,'Girl':5})
```

```python
titanic_drop.head()
```

```python
len_train=train.shape[0]
len_train
```

```python
train=titanic_drop[:len_train]
test=titanic_drop[len_train:]
```

```python
X_train=train.loc[:, 'Pclass':]
y_train=train['Survived']

X_test=test.loc[:, 'Pclass':]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

log_reg=LogisticRegression()
log_reg.fit(X_train, y_train)
svm_clf = SVC()
svm_clf.fit(X_train, y_train)
tree_clf=DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
print(log_reg.score(X_train,y_train))
print(svm_clf.score(X_train,y_train))
print(tree_clf.score(X_train,y_train))
```

```python
#↑决策树最佳，决策树的集成是随机森林，因此直接用随机森林来拟合
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

RF=RandomForestClassifier(random_state=1)
PRF=[{'n_estimators':[100,200],'max_depth':[6,8],'criterion':['gini','entropy']}]

GSRF=GridSearchCV(estimator=RF, param_grid=PRF, scoring='accuracy',cv=2)
scores_rf=cross_val_score(GSRF,X_train,y_train,scoring='accuracy',cv=5)

```

```python
model=GSRF.fit(X_train, y_train)
pred=model.predict(X_test)
output=pd.DataFrame({'PassengerId':test['PassengerId'],'Survived':pred})
output['Survived'] = output['Survived'].astype(int)
output.to_csv('submission.csv',index=False)
```
